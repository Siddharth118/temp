#Find out the total number of people based on their illness.

```python
df["illness"].value_counts()
```

```python
df["gender"].value_counts() #How many are there in male and female.
```

```python
#Taking all age groups into list data structure
age=list((df['age'].value_counts()).index)
age.sort()
```

```python
#Differentiating people based on their age group
age_count={}
for i in df['age']:
  if i in age_count.keys():
    age_count[i]+=1
  else:
    age_count[i]=1
```

```python
#Counting number of patients in each age group
ill=[]
df1=df.groupby('age')
for i in age_count.keys():
	c=df1.get_group(i)
	ill.append(c['illness'].count())
```

```python
#Pie chart to represent relation between age and illness
plt.pie(ill,labels=age,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```

#Visualize the percentage of people getting govt health Insurance due to low income, due to old age and also the percentge of people having private health insurance.

#Plot a bar chart to analyze the reduced days of activity due to iilneess based on Gender.


```python
db = df.groupby('gender')['reduced'].sum().to_frame().reset_index()
#Creating the bar chart
plt.bar(db['gender'], db['reduced'], color = ['pink', 'blue'])
#Adding the aesthetics
plt.title('Bar Chart')
plt.xlabel('gender')
plt.ylabel('reduced activity')
plt.legend(title="female\n"+"male")
#show the plot
plt.show()
```
